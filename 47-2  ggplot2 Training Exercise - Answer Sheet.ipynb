#### 1) Create a histogram of hwy mpg values using the mpg dataset

```r
pl <- ggplot(mpg, aes(x = hwy))
pl + geom_histogram(
    fill = 'red',
    alpha = 0.5,
    bins=20
)
```

#### 2) Create a bar graph per manufacturer with color fill defined by cyl count

```r
pl <- ggplot(mpg,aes(x = manufacturer))
pl + geom_bar(
    aes(fill = factor(cyl))
) + theme_minimal()
```

#### 3) Create a scatterplot of volume verses sales. Afterwards, play around with the alpha and color arguments to clarify information

```r
pl <- ggplot(data=txhousing,aes(x = sales,y=volume)) 
pl + geom_point(alpha = 0.2, color = "blue")
```

#### 4) Add a smooth fit line to the scatterplot from above. 

```r
pl + geom_point(alpha = 0.2, color = "blue") + geom_smooth(color = 'red')
```
