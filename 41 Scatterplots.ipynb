# Scatterplots

[Cheat Sheet](https://rstudio.com/wp-content/uploads/2015/03/ggplot2-cheatsheet.pdf)

Scatter plots allow us to place points that let us see possible correlations between two features of a data set. Let's see how we can create them with ggplot!

We'll use the built-in mtcars dataset:

```r
library(ggplot2)
```

```r
df <- mtcars
head(mtcars)
str(mtcars)
```

# qplot()
qplot is useful for quick and easy graphs.

```r
qplot(wt, mpg, data = mtcars)
```

## Adding a 3rd feature in qplot
We can add a third feature by adding a color gradient on each point, or by resizing each point based on their value of this 3rd feature. For example:

```r
qplot(wt,mpg,data=df,color=hp, size = hp)
```

## ggplot()
How let's see how to get more control by using ggplot()

```r
pl <- ggplot(data=df,aes(x = wt,y=mpg)) 
pl + geom_point()
```

## Adding features

```r
# Data and Aesthetics Layer
pl <- ggplot(df, aes(x = wt, y=mpg))

# Geometry
print(pl + geom_point(size = 10, alpha = 0.5))
```

```r
print(pl + geom_point(aes(size =hp)))
# works for continuous data
```

```r
# cyl is discrete so we will force it to become a discrete value.
print(pl + geom_point(aes(size = factor(cyl))))
# Ignore the warning
```

```r
# Shapes are better for discrete variables
print(pl + geom_point(
    aes(shape = factor(cyl), 
        color = factor(cyl)
       ),
    size = 5,
    alpha = 0.6)
      )
```

```r
# You can also use hex color codes into the color argument for more colors
pl2 <- pl + geom_point(
    aes(shape = factor(cyl)
    ),
    size = 5, 
    color = '#43e8d8')
print(pl2)
```

```r
# Gradient colors can be used for continuous data 
pl + geom_point(
    aes(colour = hp),
    size=4) + 
    scale_color_gradientn(colors = rainbow(5))
```
