# 2 Variable Plotting with ggplot2
Let's load in our movies dataset that comes installed with ggplot2

```r
library(ggplot2)
library(ggplot2movies)
```

```r
# Data and Aesthetics Layer
pl <- ggplot(movies, aes(x=year, y=rating))
```

```r
# Geometry layer
pl + geom_bin2d()
```

You can see that we get a "heat map" of the data. The color is based on the number of occurences per cell. We can color the graph the same way we do with the other graphs

```r
pl + geom_bin2d() + 
     scale_fill_gradient(
         high = "Red",
         low = "White")
```

You can also control the bin sizes in the heat map by passing a vector into the binwidth argument

```r
pl + geom_bin2d(
    binwidth=(c(10,1))) + 
     scale_fill_gradient(
         high = "Red",
         low = "White")
```

```r
pl + geom_hex(
   color = 'black'
) + 
    scale_fill_gradient(
        high = "red",
        low = "white") 
```

```r
pl + geom_density2d()
```

Let's mix some charts together now that we have a decent understanding of how charts work

We can use a built in dataset for Old Faithful activity for the next example

```r
str(faithful)
```

```r
m <- ggplot(faithful, aes(x = eruptions, y = waiting)) +
 geom_point()

m + geom_density_2d()
```

We can set limits on the X and Y axis so the density chart doesn't get cut off

```r
m <- ggplot(faithful, aes(x = eruptions, y = waiting)) +
 geom_point() +
 scale_color_gradientn(colors = rainbow(5))+
 xlim(0.5, 6) +
 ylim(35, 110)

m
m + geom_density_2d()
```

```r
m + stat_density_2d(alpha = 0.1, geom = "polygon")
```

```r
ggplot(faithful, aes(x = eruptions, y = waiting, fill=..level..) ) + 
      stat_density_2d( bins=11, alpha = 1, geom = "polygon") +
      scale_fill_gradientn(colours = terrain.colors(10)) +
        xlim(0.5, 6) +
        ylim(35, 110)
```
